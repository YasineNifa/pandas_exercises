```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
df = pd.read_csv(url, sep="\t")
df.head(5)
```

```python
# clean the item_price column and transform it in a float
dollarizer = lambda x:float(x[1:])
df['item_price'] = df['item_price'].apply(dollarizer)
df.info()
```

```python
# delete the duplicates in item_name and quantity and keep the first
df_filterd_1 = df.drop_duplicates(['item_name','quantity'])#keep='last'
df_filterd_1.head(5)
```

```python
# select only the products with quantity equals to 1
df[df['quantity'] == 1]
```

```python
# select only the products with item_price > 10
df[df['item_price'] > 10]
```

```python
# delete the duplicates in item_name and quantity
df_filtered_2 = df.drop_duplicates(['item_name','quantity'])
df_filtered_2.head(5)
```

```python
# select only the products with quantity equals to 1
df_filtered_2[df_filtered_2['quantity'] == 1]
```

```python
# select only the item_name and item_price columns
c = df_filtered_2[['item_name','item_price']]
c.head(5)
```

```python
# sort the values from the most to less expensive
c.sort_values('item_price',ascending=False)
```
